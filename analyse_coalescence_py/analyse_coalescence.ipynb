```python
def plot(df):
    f = plt.figure()
    ax = plt.subplot()
    ax.set_yscale('log')
    ax.plot(df.time, df["size"])
    
def plot(df_m, df_f):
    f = plt.figure()
    ax = plt.subplot()
    ax.set_yscale('log')
    ax.plot(df_m.time, df_m["size"], color = 'r')
    ax.plot(df_f.time, df_f["size"], color = 'k')

    
```
